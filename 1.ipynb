```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.9, 2.2, 3.5, 4.5, 5.7, 6.7])
y = np. array([1.1, 1.6, 2.6, 3.2, 4.0, 5.1])

plt.scatter(x, y)
```

```python
def polyModel(x, y, p):
    X = np.zeros((p+1, len(x)))
    for i in range(p+1):
        X[i] = x**i

    b_bar = np.linalg.inv(X@X.T)@X@y

    return X.T, b_bar

p = 5
X, b_bar = polyModel(x, y, p)
y_bar = X @ b_bar
error = round(sum(abs(y_bar-y)),4)
def func(x, b):
    y = np.zeros(len(x))
    for i in range(len(b)):
        y += b[i]*x**i

    return y
b = 5
x1 = np.arange(0,7,0.05)
y1 = func(x1,b_bar)

plt2 = plt.scatter(x,y,c='r')
plt1 = plt.scatter(x, y_bar,marker='*')
plt.title("p="+str(p)+",error="+str(error))
```
